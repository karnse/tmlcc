```python
import json
t = open('func_gro.json')
func_group=json.load(t)
t = open('map_topology.json')
topo = json.load(t)
```

```python
topo
```

```python
with open('train.csv') as fp:
    lines = fp.readlines()
```

```python
with open('sra_2to15.csv') as fp2:
    lines2 = fp2.readlines()
```

```python
data=[]
for line in lines:
    data.append(line.split(','))
datae=[]
for line in lines2:
    datae.append(line.split(','))
    
```

```python
cnt=0
dats=[]
for i in range(15):
    dats.append([])
for i in data[1:]:
    if i[3]=='0' or i[4]=='0' or i[5]=='0' or i[11]=='0' or i[12]=='' or i[13]=='0' or i[-2]=='inf':
        continue
    cnt+=1
    dats[-1].append(i[0])
    dats[0].append(float(i[1]))
    dats[1].append(float(i[2]))
    dats[2].append(float(i[3]))
    dats[3].append(float(i[4]))
    dats[4].append(float(i[5]))
    func = i[6].split('-')
    if func[0]=='':
        dats[5].append(-1)
        dats[6].append(-1)
    elif len(func)==1:
        dats[5].append(func_group[func[0]])
        dats[6].append(-1)
    elif len(func)==2:
        dats[5].append(func_group[func[0]])
        dats[6].append(func_group[func[1]])
    dats[7].append(int(i[7]))
    dats[8].append(int(i[8]))
    dats[9].append(int(i[9]))
    dats[10].append(topo[i[10]])
    dats[11].append(float(i[11]))
    dats[12].append(float(i[12]))
    dats[13].append(float(i[13]))
cnt
```

```python
for i in datae[1:]:
    for j in range(len(dats[0])):
        if dats[-1][j]==i[0]:
            print(i[0])
            dats[7][j]=int(i[1])
            dats[8][j]=int(i[2])
            dats[9][j]=int(i[3])
            dats[10][j]=topo[i[4]]
```

```python
import pandas as pd
pre_data=pd.DataFrame()
pre_data['MOFname']=dats[-1]
pre_data['volume [A^3]']=dats[0]
pre_data['weight [u]']=dats[1]
pre_data['surface_area [m^2/g]']=dats[2]
pre_data['void_fraction']=dats[3]
pre_data['void_volume [cm^3/g]']=dats[4]
pre_data['functional_groups_1']=dats[5]
pre_data['functional_groups_2']=dats[6]
pre_data['metal_linker']=dats[7]
pre_data['organic_linker1']=dats[8]
pre_data['organic_linker2']=dats[9]
pre_data['topology']=dats[10]
pre_data['CO2/N2_selectivity']=dats[11]
pre_data['heat_adsorption_CO2_P0.15bar_T298K [kcal/mol]']=dats[12]
pre_data['CO2_working_capacity [mL/g]']=dats[13]
```
